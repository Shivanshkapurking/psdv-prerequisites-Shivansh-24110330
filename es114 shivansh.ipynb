```python
def sets_(l1,l2):
    results=[]
    l=l1+l2
    results.append(set(l))
    s2,s3=[],[]
    for i in l1:
        if i in l2:
            s2.append(i)
        else:
            s3.append(i)
    results.extend([set(s2),set(s3)])
    return results
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
list1=[12,13,14,15,16,17]
list2=[11,13,10,20,21]
# import random
# for i in range(5):
#     list1.append(random.randrange(0,10,1))
#     list2.append(random.randrange(5,20,1))
s1,s2=set(list1),set(list2)
res=sets_(list1,list2)
print("The union is ",res[0])
print("The intersection is ",res[1])
print("The symmetric differece is",res[2])
venn=venn2([s1,s2],["A",'B'])
plt.show()
```
